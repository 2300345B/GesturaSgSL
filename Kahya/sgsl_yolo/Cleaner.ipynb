```python
import os

label_dir = r"C:\Users\Admin\Documents\GitHub\GesturaSgSL\Kahya\sgsl_yolo\dataset\labels"
max_class = 11  # You have 12 classes: 0 to 11

for subset in ["train", "val"]:
    subset_path = os.path.join(label_dir, subset)
    for file in os.listdir(subset_path):
        if file.endswith(".txt"):
            file_path = os.path.join(subset_path, file)
            with open(file_path, "r") as f:
                lines = f.readlines()

            # Keep only valid lines
            valid_lines = []
            for line in lines:
                try:
                    class_id = int(line.split()[0])
                    if class_id <= max_class:
                        valid_lines.append(line)
                    else:
                        print(f"‚ùå Removed invalid class {class_id} in {file}")
                except Exception as e:
                    print(f"‚ö†Ô∏è Skipped bad line in {file}: {line.strip()}")

            # Overwrite file with only valid lines
            with open(file_path, "w") as f:
                f.writelines(valid_lines)
```
